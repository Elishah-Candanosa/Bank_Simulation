# Funciones de extracción de datos

Una vez que acabemos la simulación, es necesario extraer los datos que la misma arroje. Esto lo podemos hacer con funciones que alteren los datos iniciales del entorno por aquellos que genere el programa, de la siguiente manera:

```python
datos_cajeros=[]
tiempo_total=0
tiempo_add=0
Gente=[]
num_Gente=0
desocupacion=[]
tamaño=[]
tiempo_caja=[]
tiempo_fila=[]
def ext_cajeros(datos_cajeros_nuevos):
  globals()['datos_cajeros']=datos_cajeros_nuevos
  return datos_cajeros_nuevos
def ext_gente(gente_total_nueva):
  globals()['Gente']=gente_total_nueva
  return gente_total_nueva
def ext_num_gente(num_gente_nueva):
  globals()['num_Gente']=num_gente_nueva
  return num_gente_nueva
def ext_desocupacion(desocupados):
  globals()['desocupacion']=desocupados
  return desocupados
def ext_tamano(tamano_nuevo):
  globals()['tamaño']=tamano_nuevo
  return tamano_nuevo
def ext_tiempo(tiempo_nuevo):
  globals()['tiempo_total']=tiempo_nuevo
  return tiempo_nuevo
def ext_tiempo_ad(tiempo_ad_nuevo):
  globals()['tiempo_add']=tiempo_ad_nuevo
  return tiempo_ad_nuevo
def ext_tiempo_caja(tiempo_caja_nuevo):
  globals()['tiempo_caja']=tiempo_caja_nuevo
  return tiempo_caja_nuevo
def ext_tiempo_fila(tiempo_fila_nuevo):
  globals()['tiempo_fila']=tiempo_fila_nuevo
  return tiempo_fila_nuevo
```

# Clases de la Simulación
La simulación se hará en el paradigma POO, por lo que debemos definir las clases que auxiliarán en la ejecución del programa:

```python
'''
Generación de las clases para la simulación
'''


#i) Clase Cuentahabiente
    #Primero, generemos las funciones que darán lugar a las entidades aleatorias con las que operará la simulación
      #Para generar identificadores diferentes para cada cliente, podemos definir una función
      #que genere "Strings aleatorias", que tendrán una longitud,por mencionar algo, de 5 caracteres:
def identificador():
    letras = string.ascii_lowercase
    result_str = ''.join(random.choice(letras) for i in range(0,5))
    return result_str
      #Luego generemos otra funcion que de manera aleatoria elija la cantidad de dinero
      #involucrada en la transacción, en función de los parámetros que el usuario indicó en el espacio de
      #Widgets:
def Dinero(Inf,Sup):
    #Usamos la librería random, con un step de 0.01
    Dinero=round(random.uniform(Inf, Sup),2) #random.uniform genera números aleatorios float, y con la función round, truncamos dicho número a las dos posiciones decimales solicitadas
    return Dinero
    #Una vez definidos los anteriores métodos, podemos aplicarlos en la clase "Cuentahabiente":
class Cuentahabiente:
    def __init__(self,DInf,DSup):
      self.identificador=identificador()
      self.dinero=Dinero(DInf,DSup)
      #Y ahora, analicemos "Cuentahabiente.dinero":
      if self.dinero<0:
        self.tipoTransaccion="Retiro"
      elif self.dinero>0:
        self.tipoTransaccion="Depósito"
      else:
        #En dado caso que Dinero() genere 0, significa que el cliente sólo "Revisó" su estatus financiero
        #Sin depositar ni retirar
        self.tipoTransaccion="Revisión"
      self.tiempo_en_fila=0


#ii) Clase coladebanco

class coladebanco:
    def __init__(self,CapFila):
      self.cola=deque(maxlen=CapFila) #definimos a la queue, y además, limitamos su tamaño al valor que proporcione el usuario
      self.intervalo=0
    #Añadir gente a la cola
    def adicion(self,DineroInf,DineroSup):
      self.cola.append(Cuentahabiente(DineroInf,DineroSup))
    # Pasar gente al cajero:
    def atender(self):
        return self.cola[0]
    #Una vez que pasó a una persona al cajero, eliminar gente:
    def eliminar(self):
      self.cola.popleft()


  #iii)Clase Cajero
class cajero:
    #Podemos también definirlo como un "queue", pero cuya máxima longitud sea 1:
    def __init__(self,numero_identificador):
      #Primero hagamos un identificador del cajero (se definirá propiamente en la simulación como tal):
      self.identificador=numero_identificador
      self.atencion=deque(maxlen=1)
      self.clientes_atendidos=0
        #Si se declara el método recibir, se aumenta en 1 "clientes_atendidos" (vea la declaración de "recibir")
      #Si cajero.atencion tiene un elemento, está ocupado (usaré booleanos para eventualmente usar checkboxes al representar a los cajeros gráficamente):
      if len(self.atencion)==1:
        self.estado=True
      #De lo contrario, está "Vacante":
      else:
        self.estado=False
      #Al activarse el método transferir, se alteran los siguientes parámetros (vea la función más abajo)
      self.dinero_depositado=0
      self.dinero_retirado=0
      self.flujo_de_caja=0
      self.depositos=0
      self.retiros=0
      self.consultas=0
      #De acuerdo a los límites de tiempo aleatorio que proporcione el usuario, definiremos los siguiente atributos
      #los cuales se irán alterando con cada segundo de simulación. Podemos definirlos inicialmente como:
      self.tiempo_total=0
      self.tiempo_instantaneo=0
      self.tiempo_restante=0

    #Métodos de la clase:
      #Recibir un cliente:
    def recibir(self,cliente):
      self.clientes_atendidos=self.clientes_atendidos+1
      self.atencion.append(cliente)
      #Terminar de atender a un cliente:
    def terminar(self):
      self.atencion.popleft()
      #Transferir dinero de un cliente:
    def transferir(self,cliente):
      if cliente.dinero>0:
        self.depositos=self.depositos+1
        self.dinero_depositado=self.dinero_depositado+cliente.dinero
      elif cliente.dinero<0:
        self.retiros=self.retiros+1
        self.dinero_retirado=self.dinero_retirado+cliente.dinero
      elif cliente.dinero==0:
        self.consultas=self.consultas+1
      self.flujo_de_caja=self.dinero_depositado+self.dinero_retirado
```

```python
def ejecucion_click(self):
  '''
  Preparación de las herramientas de Simulación
  '''
  #Primero extraemos los datos de la ventana de Ingreso de Datos:
  min=int(HorarioMinutos.value)
  seg=int(HorarioSegundos.value)
  TR=TiempoReal.value
  Caj=int(Cajeros.value)
  CapFila=int(Fila.value)
  GInf=int(AgregarInf.value)
  GSup=int(AgregarSup.value)
  DInf=float(DineroInf.value)
  DSup=float(DineroSup.value)
  TInf=int(CajeroInf.value)
  TSup=int(CajeroSup.value)
  #Como no podemos iterar sobre dos unidades de tiempo diferentes, será más fácil
  #Convertir todo a segundos y luego iterar sobre ello:
  Apertura=min*60+seg

  #Luego Podemos limpiar la ventana:
  output.clear()

  #Podemos incluso notificarle al usuario los datos con los que se ejecutará la simulación:

  print('\033[91m'+"Resumen de Parámetros de la Nueva Simulación"+'\033[0m')
  print("El Banco estará abierto por {} minutos y {} segundos.".format(min,seg))
  print("La simulación {} se llevará a cabo en Tiempo Real.".format(TR))
  print("El Banco operará con {} cajeros.".format(Caj))
  print("La Capacidad máxima de gente que tendrá la fila será: {}".format(CapFila))
  print("Cada minuto, se agregarán entre {} y {} personas a cada fila.".format(GInf,GSup))
  print("En cada Transacción Monetaria, se intercambiará entre {} y {}.".format(DInf,DSup))
  print("Cada Operación con el cajero durará entre {} y {} segundos.".format(TInf,TSup))
  print('\033[91m'+"****************************************************************************************************"+'\033[0m')




  '''
  Proceso de simulación
  '''

  print('\033[91m'+"Simulación de la Operación de un Banco:"+'\033[0m')

  #Creemos los objetos de cajeros. Como no asignaremos cada objeto a una varible, debemos diferenciarlos de alguna forma
  #La forma que se me ocurrió fue ponerlos todos en una lista, de tal suerte que permita diferenciar cada objeto en función de su posición en la lista.
  lista_cajeros=[]
  for i in range(1,Caj+1):
    lista_cajeros.append(cajero(i))

  #Y el de la cola:
  cola_simulacion=coladebanco(CapFila)

  #Y algunos elementos a usar:
  Gente_Total=[]
  num_Gente_Total=0
  cajeros_desocupados=[]
  tamaño_fila=[]
  tiempo_cajero=[]
  tiempo_fila=[]
  #Luego, creemos los widgets que representarán la información de cada objeto:
  #Para diferenciar los widgets entre sí, usemos el mismo método de diferenciación del bloque de código pasado:
  #Cola:

  print("En este momento, el Banco está:")
  Estado= widgets.Text(
            value="Por Revisar",
            disabled=False
            )
  display(Estado)
  print("Segundos de Funcionamiento del Banco:")
  Seg= widgets.IntText(
            value=0,
            disabled=False
        )
  display(Seg)
  print('\033[91m'+"******************************************************************"+'\033[0m')
  print("Fila de clientes esperando a ser atendidos:")
  Fila_Cajero =  widgets.IntSlider(
          value=len(cola_simulacion.cola),
          #Los mínimos y máximos se nos mencionan en las indicaciones del Proyecto
          step=1,
          disabled=False,
          description="Fila:",
          continuous_update=True,
          orientation='horizontal',
          readout=True,
          readout_format='d',
          layout={'width':'max-width'}
        )
  display(Fila_Cajero)

  print("Segundos adicionales, tras el tiempo de apertura, para atender a todos los clientes:")
  Seg_Ad= widgets.IntText(
            value=0,
            disabled=False
            )
  display(Seg_Ad)
  print('\033[91m'+"*************************************************************************"+'\033[0m')
        #Cajeros:
  lista_widgets_cajeros=[]
  lista_widgets_clientes=[]
  lista_widgets_depositos=[]
  lista_widgets_retiros=[]
  lista_widgets_consultas=[]
  lista_widgets_tiempo_total=[]
  lista_widgets_tiempo_momento=[]
  lista_widgets_tiempo_faltante=[]
  for j in lista_cajeros:
          #Estado del Cajero:
          Cajero=widgets.Checkbox(
            value=j.estado,
            description="Cajero "+str(j.identificador),
            disabled=False,
            indent=False
           )
          display(Cajero)
          lista_widgets_cajeros.append(Cajero)
          print('\033[91m'+"Estadísticas Totales Generales:"+'\033[0m')
          #Clientes atendidos
          print("א  Clientes Atendidos:")
          Clientes= widgets.IntText(
            value=j.clientes_atendidos,
            disabled=False
            )
          display(Clientes)
          lista_widgets_clientes.append(Clientes)
          #Depósitos
          print("א  Depósitos Realizados:")
          Depositos= widgets.IntText(
            value=j.depositos,
            disabled=False
            )
          display(Depositos)
          lista_widgets_depositos.append(Depositos)
          #Retiros
          print("א  Retiros Realizados:")
          Retiros= widgets.IntText(
            value=j.retiros,
            disabled=False
            )
          display(Retiros)
          lista_widgets_retiros.append(Retiros)
          #Consultas
          print("א  Consultas Realizadas:")
          Consultas= widgets.IntText(
            value=j.consultas,
            disabled=False
            )
          display(Consultas)
          lista_widgets_consultas.append(Consultas)
          print("..................Estadísticas para el Cliente actual en Segundos:")
          #Tiempo Total de Atención
          print("א  Tiempo Total de Atención:")
          T_A= widgets.IntText(
            value=j.tiempo_total,
            disabled=False
            )
          display(T_A)
          lista_widgets_tiempo_total.append(T_A)
          #Tiempo de Atención hasta el Momento
          print("א  Tiempo de Atención hasta el Momento:")
          T_A_M= widgets.IntText(
            value=j.tiempo_instantaneo,
            disabled=False
            )
          display(T_A_M)
          lista_widgets_tiempo_momento.append(T_A_M)
          #Tiempo de Atención Faltante
          print("א  Tiempo de Atención Faltante:")
          T_A_F= widgets.IntText(
            value=j.tiempo_restante,
            disabled=False
            )
          display(T_A_F)
          lista_widgets_tiempo_faltante.append(T_A_F)
          print('\033[91m'+"******************************************************************"+'\033[0m')



  #Primeramente, se debe ver si la Simulación será en tiempo Real o no:
  if TR=="No":
    #Entonces no se hará en Tiempo Real, y podemos iterar sobre el tiempo de Apertura
    #a la velocidad de iteración de Python:


    #Aquí inicia la simulación:
    for i in range(0,Apertura):
      #Cada iteración corresponde a "un segundo" de simulación
        #Cuando nos encontramos en estas iteraciones, se considera que el banco está Abierto:
        banco="Abierto"
        Estado.value=banco



        #Ahora, cada minuto se debe generar la cantidad de gente que se añadirá a la cola
        #a lo largo de la simulación, podemos ver a cada minuto como aquellas "i" dividibles por 60:
        if i%60==0:
          segundo_de_redefinicion=i
          gente_añadida=0
          #Gente por Minuto que se agregará a la fila cada minuto:
          GpM=random.randrange(GInf,GSup) #El step inherente a randrange es 1 porque tratamos con personas
          Gente_Total.append(GpM)
          #Ahora, para que la adición sea uniforme:
          IntervaloAdicion=math.ceil(60/GpM) #Cada cuántos segundos se agrega una persona a la cola, redondeada hacia arriba
          cola_simulacion.intervalo=IntervaloAdicion #Guardamos el intervalo en un atributo del objeto cola_simulacion.
        if (i-segundo_de_redefinicion)%IntervaloAdicion==0 and len(cola_simulacion.cola)<CapFila and banco=="Abierto": #Si la fila se llena, se deja de incluir gente en ella, al igual que cuando el banco esté cerrado
          cola_simulacion.adicion(DInf,DSup)
          gente_añadida=+1
          num_Gente_Total=num_Gente_Total+1
          Fila_Cajero.value=len(cola_simulacion.cola)

          #Como redondeamos hacia arriba, puede darse el caso que, al terminar el minuto, no se haya añadido la cantidad GpM a la cola
          #En caso de suceder dicho evento:
        if i==segundo_de_redefinicion+60 and gente_añadida!=GpM and len(cola_simulacion.cola)<CapFila and banco=="Abierto":
          gente_faltante=GpM-gente_añadida
          for j in range(0,gente_faltante):
            cola_simulacion.adicion(DInf,DSup)
            Fila_Cajero.value=len(cola_simulacion.cola)
            num_Gente_Total=num_Gente_Total+1
        #Y registramos el tamaño de la fila:
        tamaño_fila.append(len(cola_simulacion.cola))

        #Y a cada cliente en la fila le sumamos 1 en su contador tiempo_en_fila
        for j in cola_simulacion.cola:
          j.tiempo_en_fila=j.tiempo_en_fila+1

        #Ahora definamos como operará cada cajero:
          #Recibir clientes:
        for j in lista_cajeros:
          if len(j.atencion)==0 and len(cola_simulacion.cola)>0: #Como en un banco, si un cajero se libera, inmediatamente se le asignará otro cliente (si la cola tiene clientes, claro)
            #Se atiende al cliente:
            j.recibir(cola_simulacion.atender())
            j.estado=True
            #Se elimina al cliente de la fila, se registra su tamaño y se registra cuanto tiempo pasó en ella:
            tiempo_fila.append(cola_simulacion.cola[0].tiempo_en_fila)
            cola_simulacion.eliminar()
            tamaño_fila.append(len(cola_simulacion.cola))
            Fila_Cajero.value=len(cola_simulacion.cola)
            #Se determina el tiempo aleatorio de atención
            j.tiempo_total=random.randrange(TInf,TSup)
            #En el segundo i, se inició la atención al cliente
            tiempo_inicio_atencion=i
            #Se marca el widget como ocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=True #Y la usamos en la lista de Widgets


          #Atender clientes:
        for j in lista_cajeros:
          if len(j.atencion)==1:  #Es decir, si tiene un cliente
            j.tiempo_instantaneo=i-tiempo_inicio_atencion

            j.tiempo_restante=j.tiempo_total-j.tiempo_instantaneo


          #Terminar la atención al cliente:
        cajeros_desocupados_nuevos=0
        for j in lista_cajeros:
          if len(j.atencion)==1 and j.tiempo_restante==0: #Es decir, terminó el tiempo aleatorio de atención:
            j.transferir(j.atencion[0])
            j.terminar()
            Fila_Cajero.value=len(cola_simulacion.cola)
            j.estado=False
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1
            #Se marca el widget como desocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=False #Y la usamos en la lista de Widgets
            #Y se agrega el tiempo en caja a la lista correspondiente:
            tiempo_cajero.append(j.tiempo_total)

          elif len(j.atencion)==0:
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1

        cajeros_desocupados.append(cajeros_desocupados_nuevos)



        #Cuando, iterando, Python llegue aquí, habrá acabado el segundo de simulación.
        #Refresquemos los datos de los widgets que representan las estadísticas de los cajeros:
        for j in lista_cajeros:
            indice=lista_cajeros.index(j)
            #Como todos los elementos de las listas de widgets están en relación 1-1 a los elementos de la lista de cajeros, podemos
            #utilizar el mismo índice para actualizar sus valores:
            lista_widgets_clientes[indice].value=j.clientes_atendidos
            lista_widgets_depositos[indice].value=j.depositos
            lista_widgets_retiros[indice].value=j.retiros
            lista_widgets_consultas[indice].value=j.consultas
            lista_widgets_tiempo_total[indice].value=j.tiempo_total
            lista_widgets_tiempo_momento[indice].value=j.tiempo_instantaneo
            lista_widgets_tiempo_faltante[indice].value=j.tiempo_restante

            #Y actualizamos el widget de segundos de funcionamiento:
            Seg.value=i

  elif TR=="Sí":
    #Entonces se hará en tiempo Real, es básicamente el mismo método que el caso anterior
    #solo que aplicando el método sleep, que opera en segundos, que agregaremos hasta el final de cada iteración:

    #Aquí inicia la simulación:
    for i in range(0,Apertura):
      #Cada iteración corresponde a "un segundo" de simulación
      #El Banco Abre:
        banco="Abierto"
        Estado.value=banco


        #Ahora, cada minuto se debe generar la cantidad de gente que se añadirá a la cola
        #a lo largo de la simulación, podemos ver a cada minuto como aquellas "i" dividibles por 60:
        if i%60==0:
          segundo_de_redefinicion=i
          gente_añadida=0
          #Gente por Minuto que se agregará a la fila cada minuto:
          GpM=random.randrange(GInf,GSup) #El step inherente a randrange es 1 porque tratamos con personas
          Gente_Total.append(GpM)
          #Ahora, para que la adición sea uniforme:
          IntervaloAdicion=math.ceil(60/GpM) #Cada cuántos segundos se agrega una persona a la cola, redondeada hacia arriba
          cola_simulacion.intervalo=IntervaloAdicion #Guardamos el intervalo en un atributo del objeto cola_simulacion.
        if (i-segundo_de_redefinicion)%IntervaloAdicion==0 and len(cola_simulacion.cola)<CapFila and banco=="Abierto": #Si la fila se llena, se deja de incluir gente en ella de igual manera que si el banco estuviera cerrado
          cola_simulacion.adicion(DInf,DSup)
          gente_añadida=+1
          num_Gente_Total=num_Gente_Total+1
          Fila_Cajero.value=len(cola_simulacion.cola)

          #Como redondeamos hacia arriba, puede darse el caso que, al terminar el minuto, no se haya añadido la cantidad GpM a la cola
          #En caso de suceder dicho evento:
        if i==segundo_de_redefinicion+60 and gente_añadida!=GpM and len(cola_simulacion.cola)<CapFila and banco=="Abierto":
          gente_faltante=GpM-gente_añadida
          for j in range(0,gente_faltante):
            cola_simulacion.adicion(DInf,DSup)
            num_Gente_Total=num_Gente_Total+1
            Fila_Cajero.value=len(cola_simulacion.cola)

        #Y registramos el tamaño de la fila:
        tamaño_fila.append(len(cola_simulacion.cola))

        #Y a cada cliente en la fila le sumamos 1 en su contador tiempo_en_fila
        for j in cola_simulacion.cola:
          j.tiempo_en_fila=j.tiempo_en_fila+1

        #Ahora definamos como operará cada cajero:
          #Recibir clientes:
        for j in lista_cajeros:
          if len(j.atencion)==0 and len(cola_simulacion.cola)>0: #Como en un banco, si un cajero se libera, inmediatamente se le asignará otro cliente (si la cola tiene clientes, claro)
            #Se atiende al cliente:
            j.recibir(cola_simulacion.atender())
            j.estado=True
            #Se elimina al cliente de la fila, se registra su tamaño y se registra cuanto tiempo pasó en ella:
            tiempo_fila.append(cola_simulacion.cola[0].tiempo_en_fila)
            cola_simulacion.eliminar()
            tamaño_fila.append(len(cola_simulacion.cola))
            Fila_Cajero.value=len(cola_simulacion.cola)
            #Se determina el tiempo aleatorio de atención
            j.tiempo_total=random.randrange(TInf,TSup)
            #En el segundo i, se inició la atención al cliente
            tiempo_inicio_atencion=i
            #Se marca el widget como ocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=True #Y la usamos en la lista de Widgets


          #Atender clientes:
        for j in lista_cajeros:
          if len(j.atencion)==1:  #Es decir, si tiene un cliente
            j.tiempo_instantaneo=i-tiempo_inicio_atencion

            j.tiempo_restante=j.tiempo_total-j.tiempo_instantaneo


          #Terminar la atención al cliente:
        cajeros_desocupados_nuevos=0
        for j in lista_cajeros:
          if len(j.atencion)==1 and j.tiempo_restante==0: #Es decir, terminó el tiempo aleatorio de atención:
            j.transferir(j.atencion[0])
            j.terminar()
            Fila_Cajero.value=len(cola_simulacion.cola)
            j.estado=False
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1
            #Se marca el widget como desocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=False #Y la usamos en la lista de Widgets
            #Y se agrega el tiempo en caja a la lista correspondiente:
            tiempo_cajero.append(j.tiempo_total)

          elif len(j.atencion)==0:
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1

        cajeros_desocupados.append(cajeros_desocupados_nuevos)



        #Cuando, iterando, Python llegue aquí, habrá acabado el segundo de simulación.
        #Refresquemos los datos de los widgets que representan las estadísticas de los cajeros:
        for j in lista_cajeros:
            indice=lista_cajeros.index(j)
            #Como todos los elementos de las listas de widgets están en relación 1-1 a los elementos de la lista de cajeros, podemos
            #utilizar el mismo índice para actualizar sus valores:
            lista_widgets_clientes[indice].value=j.clientes_atendidos
            lista_widgets_depositos[indice].value=j.depositos
            lista_widgets_retiros[indice].value=j.retiros
            lista_widgets_consultas[indice].value=j.consultas
            lista_widgets_tiempo_total[indice].value=j.tiempo_total
            lista_widgets_tiempo_momento[indice].value=j.tiempo_instantaneo
            lista_widgets_tiempo_faltante[indice].value=j.tiempo_restante


        '''
        Aquí está la diferencia fundamental con el caso que no es en Tiempo Real:
        '''
        sleep(1) #Esta línea de código se escribe para que cada iteración dure un segundo
        #Y actualizamos el Widget de Segundos de Funcionamiento:
        Seg.value=i


  '''
  Tiempo Extra

  Ahora, aquí acaba el horario de apertura del Banco, mas existe una buena probabilidad que el banco aún tenga gente que atender
  Para ello, se sigue el siguiente código:
  '''
  #Se continúa la iteración. En este caso el límite superior para iterar lo podemos tomar arbitrariamente, siempre que nos aseguremos
  #Que se terminarán de atender a los clientes en él. En mi caso, tome el límite superior como el tiempo que tardaría un solo cajero en atender
  #a todos los clientes de la fila (mas uno, en caso que esté lleno) con el límite superior de tiempo:
  banco="Cerrado"
  Estado.value=banco
  tiempo_max=(len(cola_simulacion.cola)+1)*TSup
  if TR=="No":
    for i in range(Apertura,Apertura+tiempo_max):
    #Entonces no se hará en Tiempo Real, y podemos iterar sobre el tiempo de Apertura
    #a la velocidad de iteración de Python:
    #Aquí inicia la simulación:
      #Cada iteración corresponde a "un segundo" de simulación



        #Ahora definamos como operará cada cajero:
          #Recibir clientes:
        for j in lista_cajeros:
          if len(j.atencion)==0 and len(cola_simulacion.cola)>0: #Como en un banco, si un cajero se libera, inmediatamente se le asignará otro cliente (si la cola tiene clientes, claro)
            #Se atiende al cliente:
            j.recibir(cola_simulacion.atender())
            #Se elimina al cliente de la fila, y se registra cuanto tiempo pasó en ella:
            tiempo_fila.append(cola_simulacion.cola[0].tiempo_en_fila)
            cola_simulacion.eliminar()
            Fila_Cajero.value=len(cola_simulacion.cola)
            #Se determina el tiempo aleatorio de atención
            j.tiempo_total=random.randrange(TInf,TSup)
            #En el segundo i, se inició la atención al cliente
            tiempo_inicio_atencion=i
            #Se marca el widget como ocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=True #Y la usamos en la lista de Widgets

        #Y registramos el tamaño de la fila:
        tamaño_fila.append(len(cola_simulacion.cola))

        #Y a cada cliente en la fila le sumamos 1 en su contador tiempo_en_fila
        for j in cola_simulacion.cola:
          j.tiempo_en_fila=j.tiempo_en_fila+1

          #Atender clientes:
        for j in lista_cajeros:
          if len(j.atencion)==1:  #Es decir, si tiene un cliente
            j.tiempo_instantaneo=i-tiempo_inicio_atencion

            j.tiempo_restante=j.tiempo_total-j.tiempo_instantaneo


          #Terminar la atención al cliente:
        cajeros_desocupados_nuevos=0
        for j in lista_cajeros:
          if len(j.atencion)==1 and j.tiempo_restante==0: #Es decir, terminó el tiempo aleatorio de atención:
            j.transferir(j.atencion[0])
            j.terminar()
            Fila_Cajero.value=len(cola_simulacion.cola)
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1
            #Se marca el widget como desocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=False #Y la usamos en la lista de Widgets
            #Y se agrega el tiempo en caja a la lista correspondiente:
            tiempo_cajero.append(j.tiempo_total)

          elif len(j.atencion)==0:
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1

        cajeros_desocupados.append(cajeros_desocupados_nuevos)


        #Cuando, iterando, Python llegue aquí, habrá acabado el segundo de simulación.
        #Refresquemos los datos de los widgets que representan las estadísticas de los cajeros:
        for j in lista_cajeros:
            indice=lista_cajeros.index(j)
            #Como todos los elementos de las listas de widgets están en relación 1-1 a los elementos de la lista de cajeros, podemos
            #utilizar el mismo índice para actualizar sus valores:
            lista_widgets_clientes[indice].value=j.clientes_atendidos
            lista_widgets_depositos[indice].value=j.depositos
            lista_widgets_retiros[indice].value=j.retiros
            lista_widgets_consultas[indice].value=j.consultas
            lista_widgets_tiempo_total[indice].value=j.tiempo_total
            lista_widgets_tiempo_momento[indice].value=j.tiempo_instantaneo
            lista_widgets_tiempo_faltante[indice].value=j.tiempo_restante

        #Actualizamos el widget de tiempo extra:
        tiempo_extra=i-Apertura
        Seg_Ad.value=tiempo_extra
        #Y actualizamos el Widget de Segundos de Funcionamiento:
        Seg.value=i
        #Aquí es cuando viene la diferencia entre el banco abierto y el cerrado
        #Si la fila y los cajeros se liberan, se rompe la iteración, de lo contrario, continúa hasta hacerlo:
        reiterar="No"
        if len(cola_simulacion.cola)!=0:
          reiterar="Sí"
        for j in lista_cajeros:
            if len(j.atencion)!=0:
              reiterar="Sí"
        if reiterar=="Sí":
          continue
        elif reiterar=="No":
          tiempo_total=i
          break

  elif TR=="Sí":
    for i in range(Apertura,Apertura+tiempo_max):
    #Entonces sí se hará en Tiempo Real
    #Aquí inicia la simulación:
      #Cada iteración corresponde a "un segundo" de simulación



        #Ahora definamos como operará cada cajero:
          #Recibir clientes:
        for j in lista_cajeros:
          if len(j.atencion)==0 and len(cola_simulacion.cola)>0: #Como en un banco, si un cajero se libera, inmediatamente se le asignará otro cliente (si la cola tiene clientes, claro)
            #Se atiende al cliente:
            j.recibir(cola_simulacion.atender())
            #Se elimina al cliente de la fila,, y se registra cuanto tiempo pasó en ella:
            tiempo_fila.append(cola_simulacion.cola[0].tiempo_en_fila)
            cola_simulacion.eliminar()
            Fila_Cajero.value=len(cola_simulacion.cola)
            #Se determina el tiempo aleatorio de atención
            j.tiempo_total=random.randrange(TInf,TSup)
            #En el segundo i, se inició la atención al cliente
            tiempo_inicio_atencion=i
            #Se marca el widget como ocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=True #Y la usamos en la lista de Widgets

        #Y registramos el tamaño de la fila:
        tamaño_fila.append(len(cola_simulacion.cola))

        #Y a cada cliente en la fila le sumamos 1 en su contador tiempo_en_fila
        for j in cola_simulacion.cola:
          j.tiempo_en_fila=j.tiempo_en_fila+1

          #Atender clientes:
        for j in lista_cajeros:
          if len(j.atencion)==1:  #Es decir, si tiene un cliente
            j.tiempo_instantaneo=i-tiempo_inicio_atencion

            j.tiempo_restante=j.tiempo_total-j.tiempo_instantaneo


          #Terminar la atención al cliente:
        cajeros_desocupados_nuevos=0
        for j in lista_cajeros:
          if len(j.atencion)==1 and j.tiempo_restante==0: #Es decir, terminó el tiempo aleatorio de atención:
            j.transferir(j.atencion[0])
            j.terminar()
            Fila_Cajero.value=len(cola_simulacion.cola)
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1
            #Se marca el widget como desocupado, cada cajero está en relación 1-1
            indice=lista_cajeros.index(j)  #Tomamos la posición del cajero alterado:
            lista_widgets_cajeros[indice].value=False #Y la usamos en la lista de Widgets
            #Y se agrega el tiempo en caja a la lista correspondiente:
            tiempo_cajero.append(j.tiempo_total)

          elif len(j.atencion)==0:
            cajeros_desocupados_nuevos=cajeros_desocupados_nuevos+1

        cajeros_desocupados.append(cajeros_desocupados_nuevos)


        #Cuando, iterando, Python llegue aquí, habrá acabado el segundo de simulación.
        #Refresquemos los datos de los widgets que representan las estadísticas de los cajeros:
        for j in lista_cajeros:
            indice=lista_cajeros.index(j)
            #Como todos los elementos de las listas de widgets están en relación 1-1 a los elementos de la lista de cajeros, podemos
            #utilizar el mismo índice para actualizar sus valores:
            lista_widgets_clientes[indice].value=j.clientes_atendidos
            lista_widgets_depositos[indice].value=j.depositos
            lista_widgets_retiros[indice].value=j.retiros
            lista_widgets_consultas[indice].value=j.consultas
            lista_widgets_tiempo_total[indice].value=j.tiempo_total
            lista_widgets_tiempo_momento[indice].value=j.tiempo_instantaneo
            lista_widgets_tiempo_faltante[indice].value=j.tiempo_restante

        #Actualizamos el widget de tiempo extra:
        tiempo_extra=i-Apertura
        Seg_Ad.value=tiempo_extra
        #Y actualizamos el Widget de Segundos de Funcionamiento:
        Seg.value=i
        #Se usa sleep para que la iteración dure un segundo:
        sleep(1)
        #Aquí es cuando viene la diferencia entre el banco abierto y el cerrado
        #Si la fila y los cajeros se liberan, se rompe la iteración, de lo contrario, continúa hasta hacerlo:
        reiterar="No"
        if len(cola_simulacion.cola)!=0:
          reiterar="Sí"
        for j in lista_cajeros:
            if len(j.atencion)!=0:
              reiterar="Sí"
        if reiterar=="Sí":
          continue
        elif reiterar=="No":
          tiempo_total=i
          break








    #Aquí finalmente podemos decir que oficialmente se hizo la primera simulación. Por tanto, cambiemos el
    #estátus de la variable "estado".
  simulacion_completada()



  '''
  Finalmente, podemos reutilizar el botón nueva simulación, regresar y el botón salir, definidos con anterioridad,
  en caso que el usuario quiera realizar otra simulación
  '''
  output.clear()
  print('\033[91m'+"La Simulación ha concluido exitosamente."+'\033[0m')
  print("¡Los Resultados de la Simulación ahora son consultables desde el menú principal!")
  print(" ")
  print("*************************")
  print("Nota")
  print("En ocasiones, por la velocidad de iteración de Python, la interfaz no es realmente apreciable cuando se elige no correr la Simulación en tiempo real")
  print("Si desea apreciarla mejor, se sugiere elegir la opción de correrla en Tiempo Real.")

  navegacionejecucion=widgets.HBox([botonSim,Regresar,botonSalir])
  print("            ")
  print('\033[91m'+"*****************************************************"+'\033[0m')
  print('\033[91m'+"Menú de Navegación:"+'\033[0m')
  display(navegacionejecucion)
  #Extraemos los datos importantes producidos en la simulación:
  ext_cajeros(lista_cajeros)
  ext_gente(Gente_Total)
  ext_desocupacion(cajeros_desocupados)
  ext_tamano(tamaño_fila)
  ext_tiempo(tiempo_total)
  ext_tiempo_ad(tiempo_extra)
  ext_num_gente(num_Gente_Total)
  ext_tiempo_caja(tiempo_cajero)
  ext_tiempo_fila(tiempo_fila)
#Y llegó la hora de asociar el método con su botón:
ejecucion.on_click(ejecucion_click)
```
